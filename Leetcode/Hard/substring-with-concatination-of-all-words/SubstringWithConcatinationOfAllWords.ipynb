30. Substring with Concatenation of All Words
Hard

You are given a string s and an array of strings words of the same length. Return all starting indices of substring(s) in s that is a concatenation of each word in words exactly once, in any order, and without any intervening 

characters.

You can return the answer in any order.

 

Example 1:

Input: s = "barfoothefoobarman", words = ["foo","bar"]

Output: [0,9]

Explanation: Substrings starting at index 0 and 9 are "barfoo" and "foobar" respectively.

The output order does not matter, returning [9,0] is fine too.

Example 2:

Input: s = "wordgoodgoodgoodbestword", words = ["word","good","best","word"]

Output: []

Example 3:

Input: s = "barfoofoobarthefoobarman", words = ["bar","foo","the"]

Output: [6,9,12]
 

Constraints:

1 <= s.length <= 104

s consists of lower-case English letters.

1 <= words.length <= 5000

1 <= words[i].length <= 30

words[i] consists of lower-case English letters.

```python
from typing import List

class Solution:
    def findSubstring(self, s: str, words: List[str]) -> List[int]:
        return self.findSubword(s, words)
        
    def findSubword(self, s, words):
        result = []
        n = len(s)
        num_words = len(words)
        word_len = len(words[0])
        words_total_length = len(words) * len(words[0])
        for i in range(n - words_total_length + 1):
            if self.composite(s[i:i+words_total_length], words.copy(), words_total_length , word_len):
                result.append(i)
        return result
    
    
    def composite(self, substring, words, substring_len, word_len):
        for i in range(0, substring_len, word_len):
            slice = substring[i:i + word_len]
            if slice in words:
                words.remove(slice)
        return not words 
```
