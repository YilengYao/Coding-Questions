37. Sudoku Solver
Hard

Write a program to solve a Sudoku puzzle by filling the empty cells.

A sudoku solution must satisfy all of the following rules:

Each of the digits 1-9 must occur exactly once in each row.

Each of the digits 1-9 must occur exactly once in each column.

Each of the digits 1-9 must occur exactly once in each of the 9 3x3 sub-boxes of the grid.

The '.' character indicates empty cells.

 

Example 1:


Input: board = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]

Output: [["5","3","4","6","7","8","9","1","2"],["6","7","2","1","9","5","3","4","8"],["1","9","8","3","4","2","5","6","7"],["8","5","9","7","6","1","4","2","3"],["4","2","6","8","5","3","7","9","1"],["7","1","3","9","2","4","8","5","6"],["9","6","1","5","3","7","2","8","4"],["2","8","7","4","1","9","6","3","5"],["3","4","5","2","8","6","1","7","9"]]

Explanation: The input board is shown above and the only valid solution is shown below:


 

Constraints:

board.length == 9

board[i].length == 9

board[i][j] is a digit or '.'.

It is guaranteed that the input board has only one solution.

```python
from typing import List

class Solution:
    def solveSudoku(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        h = len(board)
        w = len(board[0])
        row_elements = {}
        column_elements = {}
        grid_elements = {}
        result = {}
        self.generateRowElements(board, h, row_elements)
        self.generateColumnElements(board, h, w, column_elements)
        
        for row_slice in range(0, h, 3):
            for column_slice in range(0, w, 3):
                grid = [row[column_slice:column_slice+3] for row in board[row_slice:row_slice+3]]
                self.generateGridElements(grid, row_slice // 3, column_slice // 3, grid_elements)
        empty_elements = self.generateEmptyElements(board, h, w)
        self.recursiveSolver(empty_elements, row_elements, column_elements, grid_elements, result)
        for key, value in result.items():
            board[key[0]][key[1]] = value
        return True
    
    def recursiveSolver(self, empty_elements, row_elements, column_elements, grid_elements, result):
        if not empty_elements:
            return True
        first_empty_element = empty_elements[0]
        candidates = row_elements[first_empty_element[0]].intersection(column_elements[first_empty_element[1]]).intersection(grid_elements[(first_empty_element[0]//3,first_empty_element[1]//3)])
        if candidates:
            for candidate in candidates:
                row_elements[first_empty_element[0]].remove(candidate)
                column_elements[first_empty_element[1]].remove(candidate)
                grid_elements[(first_empty_element[0]//3,first_empty_element[1]//3)].remove(candidate)
                if self.recursiveSolver(empty_elements[1:], row_elements, column_elements, grid_elements, result):
                    result[first_empty_element] = candidate
                    return True
                else:
                    row_elements[first_empty_element[0]].add(candidate)
                    column_elements[first_empty_element[1]].add(candidate)
                    grid_elements[(first_empty_element[0]//3,first_empty_element[1]//3)].add(candidate)
        return False
        
    def generateRowElements(self, board, board_height, row_elements):
        for row in range(board_height):
            digits = set([i for i in "123456789"])
            for element in board[row]:
                if element != ".":
                    digits.remove(element)
            row_elements[row] = digits
        
    def generateColumnElements(self, board, board_height, board_width, column_elements):
        for column in range(board_width):
            digits = set([i for i in "123456789"])
            for row in range(board_height):
                if board[row][column] != ".":
                    digits.remove(board[row][column])
            column_elements[column] = digits
            
    def generateGridElements(self, grid, row_index, column_index, grid_elements):
        digits = set([i for i in "123456789"])
        for row in grid:
            for element in row:
                if element != ".":
                    digits.remove(element)
        grid_elements[(row_index, column_index)] = digits
        
    def generateEmptyElements(self, grid, board_height, board_width):
        result = []
        for row in range(board_height):
            for column in range(board_width):
                if grid[row][column] == ".":
                    result.append((row, column))
        return result
```
