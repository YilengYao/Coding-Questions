34. Find First and Last Position of Element in Sorted Array
Medium

Given an array of integers nums sorted in non-decreasing order, find the starting and ending position of a given target value.

If target is not found in the array, return [-1, -1].

You must write an algorithm with O(log n) runtime complexity.

 

Example 1:

Input: nums = [5,7,7,8,8,10], target = 8

Output: [3,4]

Example 2:

Input: nums = [5,7,7,8,8,10], target = 6

Output: [-1,-1]

Example 3:

Input: nums = [], target = 0

Output: [-1,-1]
 

Constraints:

0 <= nums.length <= 105

-109 <= nums[i] <= 109

nums is a non-decreasing array.

-109 <= target <= 109

```python
from typing import List

class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        index = self.findNum(nums, 0, n - 1, target)
        if index == -1:
            return [-1, -1]
        return self.expand(nums, index, target)
    
    def expand(self, nums, index, target):
        n = len(nums)
        return [self.findLowMid(nums, 0, index, target), self.findHighMid(nums, index, n -1, target)]
    
    def findLowMid(self, nums, l, r, target):
        mid = (l + r)// 2
        if r - l < 3: 
            for i in range(l, r+1):
                if nums[i] == target:
                    return i
        if nums[mid] == target:
            return self.findLowMid(nums, l, mid, target)
        else:
            return self.findLowMid(nums, mid + 1, r, target)
        
        
    def findHighMid(self, nums, l, r, target):
        if r - l < 3: 
            for i in range(r, l-1 , -1):
                if nums[i] == target:
                    return i
        mid = (l + r)// 2
        if nums[mid] == target:
            return self.findHighMid(nums, mid, r, target)
        else:
            return self.findHighMid(nums, l, mid - 1, target)
        
    def findNum(self, nums, l, r, target):
        if l > r:
            return -1
        if l == r:
            return l if nums[l] == target else -1
        mid = (l + r) // 2
        
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            return self.findNum(nums, mid + 1, r, target)
        else:
            return self.findNum(nums, l, mid - 1, target)
        
```
