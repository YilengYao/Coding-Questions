You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.

 

Example 1:

Input: lists = [[1,4,5],[1,3,4],[2,6]]

Output: [1,1,2,3,4,4,5,6]

Explanation: The linked-lists are:
[
  1->4->5,
  1->3->4,
  2->6
]

merging them into one sorted list:

1->1->2->3->4->4->5->6

Example 2:

Input: lists = []

Output: []

Example 3:

Input: lists = [[]]

Output: []
 

Constraints:

k == lists.length

0 <= k <= 10^4

0 <= lists[i].length <= 500

-10^4 <= lists[i][j] <= 10^4

lists[i] is sorted in ascending order.

The sum of lists[i].length won't exceed 10^4.

Using recursion and backtracking

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if lists:
            return self.recursiveMergeKLists(lists)
        else:
            return None
        
    def recursiveMergeKLists(self, lists):
        n = len(lists)
        if n < 1:
            return []
        if n == 1:
            return lists[0]
        if n == 2:
            return self.mergeTwoLists(lists[0], lists[1])
        mid = n // 2
        return self.recursiveMergeKLists([self.recursiveMergeKLists(lists[:mid])] + [self.recursiveMergeKLists(lists[mid:])])
        
    def mergeTwoLists(self, list_one, list_two):
        head = ListNode()
        walker = head
        while list_one and list_two:
            if list_one.val <= list_two.val:
                walker.next = list_one
                list_one = list_one.next
                walker = walker.next 
            else:
                walker.next = list_two
                list_two = list_two.next
                walker = walker.next 
        if list_one:
            walker.next = list_one
        if list_two:
            walker.next = list_two
        return head.next
        
```
